## 🎯 便利な関数

```python
# ファイル一覧表示
def show_recent_files(days=7):
    """最近更新されたファイルを表示"""
    from datetime import datetime, timedelta
    import pandas as pd
    
    recent_files = []
    cutoff_time = datetime.now() - timedelta(days=days)
    
    for file_path in WORK_DIR.rglob('*'):
        if file_path.is_file() and 'backups' not in str(file_path):
            mtime = datetime.fromtimestamp(file_path.stat().st_mtime)
            if mtime > cutoff_time:
                recent_files.append({
                    'ファイル': file_path.name,
                    'パス': str(file_path.relative_to(WORK_DIR)),
                    '更新日時': mtime.strftime('%Y-%m-%d %H:%M:%S'),
                    'サイズ': f"{file_path.stat().st_size / 1024:.1f} KB"
                })
    
    df = pd.DataFrame(recent_files)
    if not df.empty:
        df = df.sort_values('更新日時', ascending=False)
        display(df)
    else:
        print("最近更新されたファイルはありません")

# 実行
show_recent_files()
```
